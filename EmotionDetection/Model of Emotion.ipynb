## Model of Emotion (Plutchik)

```python
import networkx as nx

%matplotlib inline
```

### Define Emotions

#### Basic emotions (8)

```python
emotions_list_basic = ['anger', 'anticipation', 'digust', 'fear', 'joy', 'sadness', 'surprise', 'trust']
```

#### Intensities of basic emotions

```python
emotions_list_intensities = ['serenity', 'pensiveness', 'ecstasy', 'grief', 'acceptance', 'boredom', 'admiration',
                             'loathing', 'apprehension', 'annoyance', 'terror', 'rage', 'distraction', 'interest',
                             'amazement', 'vigilance']
```

#### Combinations - Dyads

```python
emotions_list_primary_blend = {'aggressiveness': ['anger', 'anticipation'],
                               'optimism': ['anticipation', 'joy'],
                               'contempt': ['disgust', 'anger'],
                               'awe': ['fear', 'surprise'],
                               'love': ['joy', 'trust'],
                               'remorse': ['sadness', 'disgust'],
                               'disapproval': ['surprise', 'sadness'],
                               'submission': ['trust', 'fear']}
```

```python
emotions_list_secondary_blend = {'pride': ['anger', 'joy'],
                                 'hope': ['anticipation', 'trust'],
                                 'cynicism': ['disgust', 'anticipation'],
                                 'despair': ['fear', 'sadness'],
                                 'guilt': ['joy', 'fear'],
                                 'envy': ['sadness', 'anger'],
                                 'unbelief': ['surprise', 'disgust'],
                                 'curiosity': ['trust', 'surprise']}
```

```python
emotions_list_tertiary_blend = {'dominance': ['anger', 'trust'],
                                'anxiety': ['anticipation', 'fear'],
                                'morbidness': ['disgust', 'joy'],
                                'shame': ['fear', 'disgust'],
                                'outrage': ['surprise', 'anger'],
                                'sentimentality': ['trust', 'sadness'],
                                'delight': ['joy', 'surprise'],
                                'pessimism': ['sadness', 'anticipation']}
```

### Create graph / network of emotions

```python
emotions_list = {}
```

```python
emotions_list.update(emotions_list_primary_blend)
emotions_list.update(emotions_list_secondary_blend)
emotions_list.update(emotions_list_tertiary_blend)
```

```python
emotions_graph = nx.DiGraph(emotions_list)
```

```python
nx.draw_spring(emotions_graph)
emotions_graph_labels = nx.draw_networkx_labels(emotions_graph,pos=nx.spring_layout(emotions_graph))
```
